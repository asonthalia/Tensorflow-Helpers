```
import tensorflow as tf
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
# This is the numpy method
x = np.linspace(-3.0,3.0,100)
print("Printing the created numpy array:")
print(x)
print("Printing the shape of the created array:")
print(x.shape)
print("Printing the datatype of the created array:")
print(x.dtype)
```

```
# This is the Tensorflow method
x = tf.linspace(-3.0, 3.0, 100)
print("Printing the x output:")
print(x)
# This output is an operation of the tensor which contains the operation which is linspace, it contains the shape and the datatype of the graph
# The operation has been added to the default graph
```

```
# The default graph
g = tf.get_default_graph()
print("An object of a tensorflow graph is created at the memory signature:")
print(g)

# To access the contents of the default graph we can use the list traversal
print("Accessing a list of all the operations in the graph as: ")
[op.name for op in g.get_operations()]
```

```
# Get the tensor using the Tensor's name
g.get_tensor_by_name("LinSpace" + ":0")
```

```
# To compute anything you need a Tensorflow session, to create a session in Tensorflow
session = tf.Session()
print("The session object is created at the memory signature:")
print(session)
```

```
# To compute the created operation tensor you can call
print("\n[WAY 1] The session computes the output of the Tensor:")
computed_x_way1 = session.run(x)
print(computed_x_way1)
print("\n[WAY 2] The Tensor uses a session to evaluate itself:")
computed_x_way2 = x.eval(session = session)
print(computed_x_way2)
```

```
# Closing a session
session.close()
```

```
# If we want to make a session to manage a particular graph
sess = tf.Session(graph = g)
# By default it takes the default graph into account for the session
# We can create an Interactive Session which doesn't need the user to explicitly define the session again and again
sess = tf.InteractiveSession()
print("An Interative session object is therefore created with the memory signaure at:")
print(sess)
```

```
# Now we can evaluate in a interactive session without specifying the session
x.eval()
```

```
# In tensorflow we have to compute a tensor's shape so we need to call the get_shape function
print(x.get_shape().as_list())
```

```
# Creating a gaussian curve
mean = 0.0
sigma = 1.0
z = tf.exp(tf.negative(0.5 * tf.pow(((x-mean)/sigma),2))) / (sigma * tf.sqrt(2.0 * 3.142))
res = z.eval()
plt.plot(res)

# Number of values in the gaussian curve
ksize = z.get_shape().as_list()[0]
print("The number of values in the curve are: ")
print(ksize)
```

```
# Reshaping the z values to [100, 1]
z_n = tf.reshape(z, [ksize, 1])
# Transposing the gaussian vector to create a 2D gaussian
z_t = tf.reshape(z, [1, ksize])

# Creating the 2d vector
z_2d = tf.matmul(z_n, z_t)

# Displaying the 2D gaussian
plt.imshow(z_2d.eval())
```

```
# Understanding convolution
from skimage import data
img = data.camera().astype(np.float32)
print("Our camera man image is as follows:")
plt.imshow(img, cmap='gray')

print("\nThe dimensions of our image are:")
print(img.shape)
```

```
# Reshape images function - making a 2D image 4D for convolutions - Images * Width * Height * Channels
# Numpy function
img_4d = img.reshape(1, img.shape[0], img.shape[1], 1)
# Tensorflow's built in function
img_4d = tf.reshape(img, [1, img.shape[0], img.shape[1], 1])
print("The dimensions of the reshaped image are [Images * W * H * C]:")
print(img_4d.shape.as_list())
```

```
# Using our gaussian kernel to convolve the image
# Converting the kernel to be 4D
z_4d = tf.reshape(z_2d, [ksize, ksize, 1, 1])
print("The shape of the reshaped kernel is [WidthOfKernel * HeightOfKernel * Channels * NumberOfKernels] : ")
print(z_4d.shape.as_list())

# Convolving
img_4d_convolved = tf.nn.conv2d(img_4d, z_4d, strides = [1, 1, 1, 1], padding = 'SAME')
res = img_4d_convolved.eval()
print(res.shape)
```

```
# Printing the convolved image
res_img = np.squeeze(res)
plt.imshow(res_img, cmap = 'gray')

res_img2 = res[0, :, :, 0]
plt.imshow(res_img2, cmap = 'gray')
```

```
# Understanding the Gabor distribution

# Creating a sinwave of 100 values from -3 to +3 std. devns.
xs = tf.linspace(-3.0, 3.0, 100)
ys = tf.sin(xs)
print("The sine wave generated is: \n")
plt.plot(ys.eval())

# Restructuring the sinwave to make it it multiples of one, transforming the size to ksize, 1
ys = tf.reshape(ys, [ksize, 1])
print("The shape of the reshaped wave is: ")
print(ys.shape.as_list())
print("\n")

# Constructing a one's matrix
ones = tf.ones((1, ksize))
wave = tf.matmul(ys, ones)
```

```
# Plotting the multiplied wave
plt.imshow(wave.eval(), cmap = 'gray')
```

```
# Creating the gabor kernel using the gaussian kernel and a sin wave
gabor = tf.multiply(wave, z_2d)
plt.imshow(gabor.eval(), cmap = 'gray')
```

```
# Concept of placeholders - These are elements of the graph that are not known like the input and the output and can have any shape or size
# Usually the input and the output layers of the graph
img = tf.placeholder(tf.float32, shape = [None, None], name = "img")
```

```
# Continuing
print("The image placeholder's dimensions are:")
print(img.get_shape().as_list())

# Reshaping the image placeholder to have 4 dimensions
img_3d = tf.expand_dims(img, 2)
print("\nThe shape of the expanded image placeholder is:")
print(img_3d.get_shape().as_list())

# Adding the number of images placeholder col
img_4d = tf.expand_dims(img_3d, 0)
print("\nThe shape of the expanded image placeholder is:")
print(img_4d.get_shape().as_list())
```

```
# Introducing new parameters
mean = tf.placeholder(tf.float32, name = "mean")
sigma = tf.placeholder(tf.float32, name = "sigma")
ksize = tf.placeholder(tf.int32, name = "ksize")

# Doing previous operations but with placeholders
x = tf.linspace(-3.0, 3.0, 100)
z = tf.exp(tf.negative(0.5 * tf.pow(((x-mean)/sigma),2))) / (sigma * tf.sqrt(2.0 * 3.142))
z_n = tf.reshape(z, [ksize, 1])
z_t = tf.reshape(z, [1, ksize])
z_2d = tf.matmul(z_n, z_t)
ys = tf.sin(xs)
ys = tf.reshape(ys, [ksize, 1])
ones = tf.ones((1, ksize))
wave = tf.matmul(ys, ones)
gabor = tf.multiply(wave, z_2d)
gabor_4d = tf.reshape(gabor, tf.stack([ksize, ksize, 1, 1]))
convolved = tf.nn.conv2d(img_4d, gabor_4d, strides = [1, 1, 1, 1], padding = "SAME", name = "convolved")
convolved_2d = convolved[0,:,:, 0]
```

```
# Evaluate using the placeholders with feed_dict
# Image convolved with the Gabor Kernel
plt.imshow(convolved_2d.eval(feed_dict = {img: data.camera(), mean: 0.0, sigma: 1.0, ksize: 100}), cmap = "gray")
```
