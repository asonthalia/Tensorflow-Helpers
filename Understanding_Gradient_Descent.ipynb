```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx
plt.style.use('ggplot')
```

```
fig = plt.figure(figsize = (10, 6))
ax = fig.gca()
x = np.linspace(-1, 1, 200)
hz = 10
cost = np.sin(hz*x) * np.exp(-x)
ax.plot(x, cost)
ax.set_ylabel("y(x) Cost")
ax.set_xlabel("x")
ax.set_label("Curve being considered")
```

```
# Finds the difference between neighbouring gradients
gradient = np.diff(cost)

# creates the figure
fig = plt.figure(figsize=(10, 6))

# fig.gca gets or creates the current axes
ax = fig.gca()

# Creating equal spaced values
x = np.linspace(-1, 1, 200)

# Function we are defining
hz = 10
cost = np.sin(hz*x)*np.exp(-x)

# Plotting the variables
ax.plot(x, cost)
ax.set_ylabel('y(x) Cost')
ax.set_xlabel('x')

# No of iterations
n_iterations = 100

# Setting the color maps
cmap = plt.get_cmap('coolwarm')
c_norm = colors.Normalize(vmin = 0, vmax = n_iterations)
scalar_map = cmx.ScalarMappable(norm=c_norm, cmap=cmap)

# Randomly initialise a starting point on the curve
init_p = 60
# np.random.randint(len(x)*0.2, len(x)*0.8)
learning_rate = 2.5

# Range function goes till --
for iter_i in range(n_iterations):
    temp = int(init_p)
    init_p = init_p - (learning_rate * gradient[temp])
    ax.plot(x[temp], cost[temp], 'go', alpha = (iter_i + 1) / n_iterations, color = scalar_map.to_rgba(iter_i))
```
